data filtering and transformation

```python
import pandas as pd

data={
    "region":["south","north","east","west","south","north"],
    "product":["laptop","moblie","laptop","tablet","mobile","tablet"],
    "sales":[50000,30000,45000,25000,20000,35000]

}

df=pd.DataFrame(data)

print(df)
```

```python
south_sales=df[df["region"]=="south"]

print(south_sales)
```

```python
high_sales=df[df["sales"]>30000]

print(high_sales)
```

```python
sorted_df=df.sort_values("sales")

print(sorted_df)
```

```python
import pandas as pd

data = {
    "Employee_ID": [101, 102, 103, 104, 105],
    "Nmae": ["Ananthu", "Balaji", "Chiththu", "Devi", None],
    "Age": [25, 30, 35, 40, 45],
    "Salary": [50000, 60000, 70000, 80000, 90000],
}

df = pd.DataFrame(data)
print(df)

print(df.isnull())

df = df.drop_duplicates()
print(df)

df["Age"] = df["Age"].fillna(df["Age"].mean())
print(df)

df = df.dropna(subset=["Salary"])
print(df)

df = df.rename(columns={"Employee_ID": "Emp_ID"})
print(df)
```

Groupby

```python
import pandas as pd

data={
    "region":["south","north","east","west","south","north"],
    "product":["laptop","moblie","laptop","tablet","mobile","tablet"],
    "sales":[50000,30000,45000,25000,20000,35000]

}

df=pd.DataFrame(data)

region_sales=df.groupby("region")["sales"].sum()

print(region_sales)
```

```python
product_sales=df.groupby("product")["sales"].sum()

print(product_sales)
```

Aggregation

```python
print(df["sales"].sum())
```

```python
print(df["sales"].mean())
```

```python
print(df["sales"].max())
```

```python
print(df["sales"].min())
```

```python
print(df["sales"].count())
```

```python
result=df.groupby("region")["sales"].agg(
    ["sum","mean","max","min"]
)

print(result)
```
