Data Cleaning

```python
import pandas as pd

data={
    "name":["john","mary","david","sam"],
    "sales":[1000,None,1500,None]
}

df=pd.DataFrame(data)

print(df)
```

```python
print(df.isnull())
```

```python
clean_df = df.dropna()

print(clean_df)
```

```python
df.fillna(0,inplace=True)

print(df)
```

```python
import pandas as pd

Data={
    "marks":[50,60,70,None]
}

df=pd.DataFrame(Data)

mean_value=df["marks"].mean()

df["marks"].fillna(mean_value,inplace=True)

print(df)
```

Handling Duplicates

```python
import pandas as pd

data={
    "name":["john","mary","john"],
    "sales":[1000,15000,1000]
}

df=pd.DataFrame(data)

print(df.duplicated())
```

```python
import pandas as pd

data={
    "name":["john","mary","john"],
    "sales":[1000,15000,1000]
}

df=pd.DataFrame(data)

df=df.drop_duplicates()

print(df)
```

Rename Columns

```python
import pandas as pd

data={
    "emp_nm":["john","mary"],
    "emp_sal":[25000,30000]
}

df=pd.DataFrame(data)

df.rename(columns={
    "emp_nm":"Empolyee name",
    "emp_sal":"Empolyee salary"
},    inplace=True)

print(df)
```
