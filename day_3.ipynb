Average Calclator

```python
def calculate_average(a, b, c):
    return (a + b + c) / 3

try:
    a = int(input("Enter 1st number: "))
    b = int(input("Enter 2nd number: "))
    c = int(input("Enter 3rd number: "))

    print("calculate_average =", calculate_average(a, b, c))

except ValueError:
    print("Invalid Input")
```

Practice 1:Student score Analysis

```python
scores=[78,85,92,67,88]

total=sum(scores)
average=total/len(scores)
highest=max(scores)
lowest=min(scores)

print("student scores:",scores)
print("total marks:",total)
print("average marks:",average)
print("highest mark:",highest)
print("lowest mark:",lowest)
```

Practice 2:Sales Array Calculations

```python
sales=[12000,15000,18000,14000,20000]

total_sales=sum(sales)
average_sales=total_sales/len(sales)
highest_sales=max(sales)
lowest_sales=min(sales)

print("monthly sales:",sales)
print("total sales:",total_sales)
print("average sales:",average_sales)
print("highest sales:",highest_sales)
print("lowest sales:",lowest_sales)
```
