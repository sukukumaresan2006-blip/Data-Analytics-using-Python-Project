E-commerce sales Analysis

```python
from google.colab import files
import pandas as pd
import matplotlib.pyplot as plt

uploaded=files.upload()

df=pd.read_csv("100 Sales Records.csv")

print(df)
```

```python
print(df.isnull().sum())
```

```python
df=df.drop_duplicates()

print(df)
```

```python
total_sales=df["Total Cost"].sum()

print("Total sales:",total_sales)
```

```python
product_sales=df.groupby("Item Type")["Total Revenue"].sum()

print(product_sales)
```

```python
product_sales.plot(kind='bar')

plt.title("product sales Analysis")

plt.xlabel("products")

plt.ylabel("sales")

plt.show()
```
