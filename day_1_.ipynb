Student Marks Calculator

```python
mark1=int(input("enter mark 1:"))
mark2=int(input("enter mark 2:"))
mark3=int(input("enter mark 3:"))

total=mark1+mark2+mark3
average=total/3

print("total:",total)
print("average:",average)

if average>=50:
  print("pass")
else:
  print("fail")
```

Expense Tracker

```python
expenses=[100,250,300,150]

total=sum(expenses)

print("expenses:",expenses)
print("total expenses:",total)
```

Grading System

```python
mark1 = int(input("Enter Mark 1: "))
mark2 = int(input("Enter Mark 2: "))
mark3 = int(input("Enter Mark 3: "))

total = mark1 + mark2 + mark3
Average = total / 3

print("Total:", total)
print("Average:", Average)

if Average >= 90:
    print("Grade: A")
elif Average >= 50:
    print("Grade: B")
else:
    print("Grade: C")
```
